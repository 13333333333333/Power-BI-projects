```python
import numpy as np
import pandas as pd

df = pd.read_csv("data/sla-sale-sites.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# dropping the redundant cols
df = df.drop(df.columns[[6,7,10]],axis=1)
```

```python
df.head()
```

```python
# replace all object 'NA' values with NaN
df.replace({'na': pd.NA, '': pd.NA}, inplace=True)
```

```python
# dropping the na values
df.dropna(inplace=True)
```

```python
df['date_of_award'] = pd.to_datetime(df['date_of_award'])
df['date_of_launch'] = pd.to_datetime(df['date_of_launch'])
df['date_of_tender_closing_or_auction'] = pd.to_datetime(df['date_of_tender_closing_or_auction'])
```

```python
df.head()
```

```python
df.to_csv("data/cleaned_sla.csv",index=False)
```
