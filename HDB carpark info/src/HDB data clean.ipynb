```python
import pandas as pd
```

```python
df = pd.read_csv("../data/hdb-carpark-information.csv")
df.head()
```

```python
df.describe()
```

```python
# getting rid of redundant naming
df['car_park_type'] = df['car_park_type'].str.replace('CAR PARK', '')
df['type_of_parking_system'] = df['type_of_parking_system'].str.replace('PARKING', '')
```

```python
# appending Singapore in front of the addresses
df["address"] = df["address"] + ", Singapore"
```

```python
df.to_csv("../data/hdb-carpark-information-edited.csv")
```
